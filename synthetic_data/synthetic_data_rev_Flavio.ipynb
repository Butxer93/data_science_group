```python
import pandas as pd
import random
from faker import Faker
import numpy as np

fake = Faker("es_ES")

def generar_vehiculos(n=1000):
    tipos_combustible = ["Gasolina", "Diésel", "Eléctrico", "Híbrido"]
    tipos_vehiculo = ["Camión", "Bus", "Turismo", "Furgoneta"]

    vehiculos = []
    for _ in range(n):
        tipo_combustible = random.choice(tipos_combustible)
        

        if tipo_combustible in ["Gasolina", "Diésel", "Híbrido"]:
            consumo_l = round(random.uniform(3, 30), 1)
        else:
            consumo_l = np.nan
        
        if tipo_combustible in ["Eléctrico", "Híbrido"]:
            consumo_kwh = round(random.uniform(10, 30), 1)
        else:
            consumo_kwh = np.nan
        
        vehiculos.append({
            "id_vehiculo": fake.license_plate(),
            "tipo_vehiculo": random.choice(tipos_vehiculo),
            "tipo_combustible": tipo_combustible,
            "peso_vacio_kg": random.randint(900, 12000),
            "altura_m": round(random.uniform(1.3, 4.5), 2),
            "consumo_medio_l_100km": consumo_l,
            "consumo_medio_kwh_100km": consumo_kwh
        })
    return pd.DataFrame(vehiculos)
```

```python
generar_vehiculos()
```

```python
def generar_rutas(n=1000):
    rutas = []
    for _ in range(n):
        inicio_ciudad, fin_ciudad = fake.city(), fake.city()
        inicio_lat, inicio_lon = round(fake.latitude(), 6), round(fake.longitude(), 6)
        fin_lat, fin_lon = round(fake.latitude(), 6), round(fake.longitude(), 6)

        paradas = [ (fake.city(), round(fake.latitude(), 6), round(fake.longitude(), 6)) for _ in range(random.randint(0,3))]

        ruta = {
            "id_vehiculo": fake.license_plate(),
            "punto_inicio_ciudad": inicio_ciudad,
            "punto_inicio_lat": inicio_lat,
            "punto_inicio_lon": inicio_lon,
            "punto_fin_ciudad": fin_ciudad,
            "punto_fin_lat": fin_lat,
            "punto_fin_lon": fin_lon,
            "evitar_peajes": random.choice([True, False])
        }

        for i, (c, lat, lon) in enumerate(paradas, start=1):
            ruta[f"parada{i}_ciudad"] = c
            ruta[f"parada{i}_lat"] = lat
            ruta[f"parada{i}_lon"] = lon

        rutas.append(ruta)

    return pd.DataFrame(rutas)
```

```python
def generar_clientes(n=1000):
    clientes = []
    for _ in range(n):
        clientes.append({
            "id_vehiculo": fake.license_plate(),
            "id_cliente": fake.uuid4(),
            "nombre": fake.name(),
            "email": fake.email(),
            "telefono": fake.phone_number(),
            "empresa": fake.company(),
            "fecha_registro": fake.date_this_decade()
        })
    return pd.DataFrame(clientes)
```
